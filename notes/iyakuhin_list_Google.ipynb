```python
import pandas as pd
import jaconv
```

```python
df = pd.read_csv('y.csv', encoding='CP932', header=None)
```

```python
df.head()
```

```python
df = df.loc[:, [4, 6]]
columns = ['name', 'furigana']
df.columns = columns
```

```python
df['furigana'] = df.apply(lambda x: jaconv.han2zen(x['furigana']), axis=1)
df['furigana'] = df.apply(lambda x: jaconv.kata2hira(x['furigana']), axis=1)
```

```python
df['alpha'] = df.apply(lambda x: jaconv.kana2alphabet(x['furigana']), axis=1)
```

```python
df
```

```python
df['database'] = df['name'].str.cat([df['furigana'], df['alpha']], sep='///')
```

```python
df['database'].to_csv('drug_list.csv')
```
